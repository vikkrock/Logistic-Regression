## **Importing the Dependencies**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## **Data Collection and Preprocessing**

```python
heart=pd.read_csv('/content/heart_disease_data.csv')
```

```python
#1st 5 rows in dataset
heart.head()
```

```python
#print last 5 rows
heart.tail()
```

```python
# number of rows and columns in the dataset
heart.shape
```

```python
# getting info about data
heart.info()
```

```python
# check missing value
heart.isnull().sum()
```

```python
#statistical measures about the data
heart.describe()
```

```python
# checking the distribution of Target Variable
heart['target'].value_counts()
```

**0--> Defective Heart**    
**1--> Healthy Heart**

### Splitting the Feature and Target

```python
X = heart.drop(columns='target',axis=1)
Y = heart['target']
```

```python
print(X)
```

```python
print(Y)
```

### Splitting the Data into Training Data & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

### Model Training

Logistic Regression

```python
model=LogisticRegression()
```

```python
# training the logistic regression model with training data
model.fit(X_train,Y_train)
```

### Model Evaluation


Accuracy Score

```python
#accuracy  on Training Data
X_train_predict = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predict, Y_train)
```

```python
print('Accuracy on Training Data : ', training_data_accuracy)
```

```python
# Accuracy of Test Data
X_test_predict = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_predict, Y_test)
```

```python
print('Accuracy on Test Data: ',test_data_accuracy)
```

### Building Predictive System

```python
input_data = (44,1,2,130,233,0,1,179,1,0.4,2,0,2)

# change the input data to numpy array
input_data_as_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for only instance
input_data_reshape= input_data_as_array.reshape(1,-1)

prediction = model.predict(input_data_reshape)

print(prediction)
```

```python
if prediction[0]==0:
  print("The Person don't have Heart disease" )
else:
  print("The Person have Heart Disease")
```
